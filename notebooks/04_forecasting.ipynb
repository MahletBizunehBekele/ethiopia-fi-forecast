```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/raw/ethiopia_fi_unified_data.csv")

df["observation_date"] = pd.to_datetime(
    df["observation_date"],
    errors="coerce"
)
```

```python
access = df[
    df["indicator_code"]=="ACC_OWNERSHIP"
].copy()

access["observation_date"] = pd.to_datetime(access["observation_date"])

access = access.sort_values("observation_date")
```

```python
years = [2025,2026,2027]

forecast = pd.DataFrame({
    "Year":years,
    "Forecast":[52,55,58]
})

forecast
```

```python
plt.figure(figsize=(8,4))

plt.plot(
    access["observation_date"],
    access["value_numeric"],
    marker="o",
    label="Historical"
)

plt.plot(
    pd.to_datetime(forecast["Year"],format="%Y"),
    forecast["Forecast"],
    marker="o",
    linestyle="--",
    label="Forecast"
)

plt.legend()

plt.grid()

plt.title("Forecast of Account Ownership")

plt.show()
```

```python
forecast["Optimistic"] = forecast["Forecast"] + 3
forecast["Pessimistic"] = forecast["Forecast"] - 3

forecast
```

## Interpretation

The forecast assumes continued expansion of digital financial services.

The optimistic scenario assumes faster adoption following additional mobile money expansion.

The pessimistic scenario assumes slower adoption due to economic or regulatory constraints.
